```python

# Accessing the csv file
filename = 'earthquakedata.csv'
with open(filename, encoding='utf8') as f:
  reader = csv.reader(f)
  header_row = next(reader)

  ## Printing each column name and its index number
  # for index, column_name in enumerate(header_row):
  #    print(index, column_name)
  ## Notes: 1-time, 2-latitude, 3-longitude, 5-mag, 14-place

  # Getting the necessary lists of values
  mags, lats, lons, hover_texts = [], [], [], []
  for row in reader:
    mag_min = 5.5
    if float(row[5]) >= mag_min:
        try:
            mag = float(row[5])
            lat = float(row[2])
            lon = float(row[3])
            hover_this = f"Mag: {mag}/ {row[14]}/ Date (Y-m-d): {row[1][:10]}"
        except ValueError:
            print(f"Missing data on {row[1]}.")
        else:
            mags.append(mag)
            lats.append(lat)
            lons.append(lon)
            hover_texts.append(hover_this)

# Customizing the map to make it more accessible to see
data = [{
 'type':'scattergeo',
 'lon': lons,
 'lat': lats,
 'text': hover_texts,
 'marker':{
  'size': [4*mag if mag>=7 else 2*mag for mag in mags],
  'color': mags,
  'colorscale': 'rdylgn',
  'reversescale': True,
  'colorbar': {'title':'Magnitude'},
 },
}]

my_title = dict(text=f"Earthquakes with Magnitude {mag_min} and Above, Since January 2000<br />Total Number of Events = {len(mags)} (Data from USGS)", font=dict(size=30))
my_layout = Layout(title=my_title, title_x=0.5)

# Producing the Map of Earthquakes Distribution
fig = {'data': data, 'layout': my_layout}
offline.plot(fig, filename='Global_EarthQuakes_Since_2000.html')
```
